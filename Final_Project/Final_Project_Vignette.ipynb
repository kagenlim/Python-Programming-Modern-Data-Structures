For students who seek to pursue a college education in Singapore, only a few have access to a liberal arts education. This means that very early on -- potentially even before they apply for college -- students need to make decisions on what their college majors might be. This is a relatively consequential decision to make. This project aims to provide a package in Python for Singaporean Pre-University students (i.e., roughly 17-18 year old students) to help them make informed, data-driven decisions on their college majors. Specifically, through the `data.gov.sg` API, this package help students, at various stages of preparedness/readiness to make decisions on their future. 

This is done by providing using functions, visualizations and easily accessible understandable tabular information, based on data from two ongoing surveys that are published on the data.gov.sg API: the a) Annual Graduate Employment Survey (done on all six autonomous universities in Singapore - Nanyang Technological University(NTU), National University of Singapore (NUS), Singapore Management University (SMU), Singapore Institute of Technology (SIT), Singapore University of Social Sciences (SUSS), and Singapore University of Technology and Design (SUTD)) and b) the Quarterly Job Vacancy Rate by Industry and Occupational Group.

For students who are most unprepared, the `sg.explore_study` function can help them discover what they might like. All they need to do is to key in one of 9 fields and they can begin exploring the full range of possibilities that the higher education scene in Singapore has to offer.

    - sg.Business
    - sg.Engineering
    - sg.Computing
    - sg.Education
    - sg.Law
    - sg.Humanities_SocialSci
    - sg.Healthcare
    - sg.Science
    - sg.Design_Art

```python
sg.explore_study(sg.Business)[0:5]
```

For students who are a bit more sure of their intended course of study, and the university they would like to study at, the following functions `get_employed`, `see_employ`, `get_paid` and `see_paid` are meant to help them dive deeper into what they might like to study. This would also give a realistic sense of how competitive the field is, post-graduation, and the kind of salary they can expect to have after graduation.

```python
sg.get_employed(sg.Engineering, sg.NUS)[0:5]
```

```python
sg.see_employ(sg.Engineering, sg.SUTD)
```

```python
sg.get_paid(sg.Law, sg.SMU)[0:5]
```

```python
sg.see_paid(sg.Law, sg.SMU)
```

For the most industry-minded, `explore_opening` and `see_opening` are meant to help them with an awareness of the opportunities that lie in particular sectors. This might help inform their decisions on what to study in university/college too.

The full list of sectors can be found in the `help(sg)` documentation below.

```python
sg.explore_opening('community, social and personal services').tail(10)
```

```python
sg.see_opening('construction')
```
