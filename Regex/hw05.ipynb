### a) Length of Narrative

```python
import pandas as pd
import re 

emergency = pd.read_csv("emergency.csv", header = None)
emergency.head()
```

```python
emergency_list = emergency.iloc[:,0]
emergency_list
```

```python
longest_string = max(emergency_list, key=len)
print(longest_string)
#longest description
```

```python
shortest_string = min(emergency_list, key=len)
print(shortest_string)
#shortest description
```

### b) Fractures

```python
fracture_s = [string for string in emergency_list if re.search(r'(FRACT(URE|URES|URED))', string)]
```

```python
fracture_fx = [string for string in emergency_list if re.search(r'FX', string)]
fracture_fx
```

```python
print(f"Total Number of fracture searches: " + str(len(fracture_s)))
print(f"Total Number of fx searches: "+ str(len(fracture_fx)))
print(f"Total Number of fracture narratives: " +str(len(fracture_s) + len(fracture_fx)))
#fractured narratives
```

### c) Body Parts

```python
all_fractures = fracture_fx + fracture_s
```

```python
parts_fx1 = [re.findall(r'\w+(?=\s+FX)', parts) for parts in all_fractures] #word that comes before FX
parts_fx1
```

```python
parts_fx2 = [re.findall(r'FX\s+((?:\w+(?:\s+|$)))', parts) for parts in all_fractures] #word that comes after FX
parts_fx2
```

```python
parts_fxcomb = parts_fx1 + parts_fx2

parts_fxall = list(filter(None, parts_fxcomb))
len(parts_fxall)
```

```python
parts_frac1 = [re.findall(r'\w+(?=\s+FRACT[A-Z]+)', parts) for parts in all_fractures] #word that comes before fractures
parts_frac1
```

```python
parts_frac2 = [re.findall(r'FRACT[A-Z]+\s+((?:\w+(?:\s+|$)))', parts) for parts in all_fractures] #word that comes after fractures
parts_frac2
```

```python
parts_fracomb = parts_frac1 + parts_frac2

parts_fracall = list(filter(None, parts_fracomb))
len(parts_fracall)
```

```python
parts_all = parts_fxall + parts_fracall
len(parts_all) 
```

```python
def get_unique_parts(parts): #getting unique parts
    unique = []
    for part in parts:
        if part not in unique:
            unique.append(part)
    return unique
```

```python
unique_parts = get_unique_parts(parts_all)

print(len(unique_parts))
unique_parts #list of unique parts, as a check
```

```python
parts_count = pd.DataFrame(parts_all).iloc[:,0]

parts_count.value_counts().head(8) #a selection of fractured body parts
```

```python
parts_tograph = parts_count.value_counts().head(8).drop(['DX', 'AND', 'RADIUS']) #removing irrelevant names
```

```python
import matplotlib.pyplot as plt

plt.bar(parts_tograph.index,parts_tograph) 
plt.xlabel('Body Part')
plt.ylabel('Fractured Count')
plt.show()
```

### d) Age and Gender

```python
#trying to capture age and gender in this section, in the whole list of injuries, emergency_list
[x for x in emergency_list if re.search(r'(\d+)\s?(YO|MO)(M|F)', x)] #age consistently comes before (YO|MO), with or without a space in between
```

```python
capture_groups = [re.findall('(\d+)\s?(YO|MO)(M|F)', x) for x in emergency_list]

capture_groups
```

```python
capture_grp_series = pd.Series(capture_groups).str.get(0) #removing outer square brackets

capture_grp_series
```

```python
df = capture_grp_series.apply(pd.Series)
df.columns=['Age','Units', 'Gender']

df = df.dropna() #to remove empty rows. These rows have no information about age or gender
df
```

```python
MO_only = df[df['Units'] == 'MO']

MO_only
```

```python
import numpy as np

MO_only_age = [int(x) for x in np.array(MO_only['Age'])] #convert to integer

MO_only_age
```

```python
divisor = 12

MO_to_YO = [x/12 for x in MO_only_age] #convert to years, through dividing by 12

MO_to_YO
```

```python
n = MO_only.columns[0]

MO_only.drop(n, axis = 1, inplace = True)

MO_only[n] = MO_to_YO #replacing Age column with transfromed MO_to_YO column

MO_only
```

```python
#shifting Age to first column  

cols = MO_only.columns.tolist()
cols.insert(0, cols.pop(cols.index('Age')))

MO_only = MO_only.reindex(columns= cols)
MO_only
```

```python
YO_only = df[df['Units'] == 'YO']

YO_only
```

```python
YO_only_age = [int(x) for x in np.array(YO_only['Age'])] #convert to integer

YO_only_age
```

```python
m = YO_only.columns[0]

YO_only.drop(n, axis = 1, inplace = True)

YO_only[m] = YO_only_age #replacing Age with transformed columns

YO_only
```

```python
#shifting Age to first column   

cols = YO_only.columns.tolist()
cols.insert(0, cols.pop(cols.index('Age')))

YO_only = YO_only.reindex(columns= cols)
YO_only
```

```python
#concatenating our two transformed data frames
all_preprocessed = pd.concat([MO_only, YO_only], axis = 0)

all_preprocessed
#shape of all_preprocessed is equivalent to df above, shows that data cleaning did not accidentally delete rows
```

```python
#variable for Gender 

Gender = all_preprocessed['Gender']

#variable for Age

Age = all_preprocessed['Age']
```

```python
male_only = all_preprocessed[Gender == 'M']

len(male_only) #Number of Males
```

```python
female_only = all_preprocessed[Gender == 'F']

len(female_only) #Number of Females
```

```python
plt.plot(male_only.groupby(male_only.Age).count(), color = "red", label = "male")
plt.plot(female_only.groupby(female_only.Age).count(), color = "black", label = "female")
plt.title('Number of Injuries by Age, for Male and Female')
plt.xlabel('Age')
plt.ylabel('Number of Injuries')
plt.text(105, 4, 'male')
plt.text(88, 20, 'female')
plt.show()
```
