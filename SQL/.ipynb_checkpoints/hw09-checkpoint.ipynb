### c) What proportion of the devils (in `devil_type`) are male?

```python
%sql SELECT COUNT(devil_type) FROM devilappearance
```

```python
%sql SELECT COUNT(devil_type) FROM devilappearance WHERE devil_type IN ('Male')
```

```python
%%sql

##combining logic of the two above in nested subquery. 

SELECT 
    devil_type, 
    (COUNT(devil_type) / (SELECT COUNT(devil_type) FROM devilappearance) * 100) AS Proportion_Male
FROM devilappearance
WHERE devil_type IN ('Male') #I am only counting values for the devil_type that explicitly states 'Male', and nothing else
```

## 3. The trial

Let's take a look at the information on the `trial`.

```python
%sql DESCRIBE trial
```

### a) What are the average and maximum numbers of male and female accusers?

```python
%%sql

SELECT 
    AVG(male_accusers), 
    AVG(female_accusers), 
    MAX(male_accusers), 
    MAX(female_accusers)
FROM trial
```

### b) Count the number of `sentence`s by sentence type. List them in a table (in descending order), excluding missing values. Rename the column headings to something sensible.

```python
%%sql

SELECT sentence AS Sentence_Type, COUNT(sentence) AS Number_of_Sentences
FROM trial
WHERE sentence IS NOT NULL #remove missing values
GROUP BY Sentence_Type
ORDER BY Number_of_Sentences DESC
```

### c) Do the number of accusers matter for the `verdict`? Compare the average number of accusers by the type of verdict. Again make sure the table is sorted and the headings make sense.

```python
%%sql

SELECT 
    verdict AS Verdict_Outcome,
    AVG(male_accusers + female_accusers) AS Average_Number_of_Accusers,
    Count(verdict) AS Number_of_Verdicts
FROM trial
GROUP BY Verdict_Outcome
ORDER BY Average_Number_of_Accusers DESC
```

The `Average_Number_of_Accusers` does seem to matter for the `Guilty` outcome -- on average, there are around $5.80$ accusers for that `Verdict_Outcome`. On average, for the `Not Guilty` outcome, there are less `Average_Number_of_Accusers`, around $4.56$ instead. 

However, it seems like at low figures of `Average_Number_of_Accusers`, on average, the `Half Guilty` or `Not Proven` `Verdict_Outcome`s are far more likely, at around $0.286$ or $1.0$ accuser(s) required respectively. For some sense of proportion, I generated `Number_of_Verdicts` as an additional check. Our understanding of the the `Half Guilty` or `Not Proven` `Verdict_Outcome`s come from just 7 and 11 cases out of the whole sample, respectively, and this may cast some uncertainty about the conclusions we can draw about them.

## 4. Tortured Truth

```python
%sql DESCRIBE confession
```

### a) Left join the `trial` and `confession` tables. For what share of trials does the database record confessions? Create a results table with the number of all trials, the number of confessions, and the share of trials with confessions recorded.

```python
%%sql

SELECT
    (SELECT COUNT(*) FROM trial) AS Number_of_all_trials,
    COUNT(confessionid) AS Number_of_all_confessions,
    COUNT(confessionid) / (SELECT COUNT(*) FROM trial) AS Share_of_trials_with_confessions, #using this as denominator instead of COUNT(*), as some trials have multiple confessions
    COUNT(*) AS Left_Joined_Number
FROM trial
LEFT JOIN confession
ON trial.trialref = confession.trialref
```

### b) Only a small number of trials have records of torture. Is there a higher share of confessions among trials with records of torture than trials without such record? Hint: You will need to merge on the `confession` table.

```python
%sql DESCRIBE torture
```

```python
%%sql

SELECT 
    COUNT(*) AS Total_confessions_with_torture,
    ((SELECT COUNT(*) FROM confession) - COUNT(*)) AS Total_Confessions_without_torture
FROM torture
INNER JOIN confession #to get the intersection only#
ON torture.trialref = confession.trialref
```

There is a smaller share of `Total_confessions_with_torture`, with $40$ such cases, compared to `Total_Confessions_without_torture`, with $901$ such cases.
